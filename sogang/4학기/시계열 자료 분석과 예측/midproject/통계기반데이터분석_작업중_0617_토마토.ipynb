```python
## timelag 위해 수집일 +1 컬럼에 착화 후 수확 기간 셋팅 
#토마토: 7주 
# 방울토마토 : 42일 
# 참고 : http://m.rda.go.kr/mobile/img/video/pdf/video_17.pdf (방울토마토 재배기술) 
# 3. 수확
# 방울토마토는 수정 후 3~5 일이면 착과되기 시작하고, 30 일 후에는 과실비대가
# 거의 완성됩니다. 수정 후 저온기에는 45~50 일, 고온기에는 35~40 일이면
# 완숙과를 수확할 수 있습니다. 

#tmtDfnew["수집일+1"] = tmtDfnew["수집일"] + pd.DateOffset(days=49)
#minTmtDfnew["수집일+1"] = minTmtDfnew["수집일"] + pd.DateOffset(days=42)
```

```python
def setOutDay (arg1, arg2):
  Dfnew = arg1.copy()
  DfOut = arg2.copy()
  ## 1달 일치 건 찾기 
  Dfnew["최종출하일"] = ""
  Dfnew["총 출하량(kg)"] = ""
  for i, newRow in Dfnew.iterrows():
    day = newRow['수집일+1']
    for j, outRow in DfOut.iterrows(): 
      outDay = outRow['수집일']
      outKg = outRow['총 출하량(kg)']
      if (day == outDay) : 
        Dfnew.loc[i, ['최종출하일']] = outDay
        Dfnew.loc[i, ['총 출하량(kg)']] = outKg
  ## pap : 6건       

  ## 일치건이 없을경우 이후 일주일 
  for i, newRow in Dfnew.iterrows():
    if str(newRow['최종출하일']) == "":
      for k in range (1,8):
        day = newRow['수집일+1']+ pd.DateOffset(days=k)
        for j, outRow in DfOut.iterrows(): 
          outDay = outRow['수집일']
          outKg = outRow['총 출하량(kg)']
          if day == outDay : 
            Dfnew.loc[i, ['최종출하일']] = outDay
            Dfnew.loc[i, ['총 출하량(kg)']] = outKg


  ## 일치건이 없을경우 이전 일주일 
  for i, newRow in Dfnew.iterrows():
    if str(newRow['최종출하일']) == "":
      for k in range (1,8):
        day = newRow['수집일+1']+ pd.DateOffset(days=k*-1)
        for j, outRow in DfOut.iterrows(): 
          outDay = outRow['수집일']
          outKg = outRow['총 출하량(kg)']
          if day == outDay : 
            Dfnew.loc[i, ['최종출하일']] = outDay
            Dfnew.loc[i, ['총 출하량(kg)']] = outKg

  return Dfnew, DfOut
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift

#X, y = make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=0.7, random_state=0)

from sklearn.cluster import estimate_bandwidth

bandwidth = estimate_bandwidth(tmt_scaled_df)
print('bandwidth 값:', round(bandwidth,3))
best_bandwidth = round(bandwidth,3)
#best_bandwidth = 0.9

meanshift= MeanShift(bandwidth=best_bandwidth)
cluster_labels = meanshift.fit_predict(tmt_scaled_df)
print('cluster labels 유형:', np.unique(cluster_labels))
```

```python
tmtLiveSmallX = tmtDfLive[['생장길이(cm)', '화방높이(cm)', '줄기굵기(mm)','엽장(cm)', '엽폭(cm)', '엽수(개)', '개화군(점)', '착과군(점)']]
tmtLiveSmallY = tmtDfLive[['열매수(개)']]
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

min_max_scaler = MinMaxScaler()
fitted = min_max_scaler.fit(tmtLiveSmallX)
print(fitted.data_max_)

tmt_scaled_df_small = min_max_scaler.transform(tmtLiveSmallX)
tmt_scaled_df_small = pd.DataFrame(tmt_scaled_df, columns=tmtXDataset.columns, index=list(tmtLiveSmallX.index.values))
print(tmt_scaled_df_small.head())
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift

#X, y = make_blobs(n_samples=200, n_features=2, centers=3, cluster_std=0.7, random_state=0)

from sklearn.cluster import estimate_bandwidth

bandwidth = estimate_bandwidth(tmt_scaled_df_small)
print('bandwidth 값:', round(bandwidth,3))
best_bandwidth = round(bandwidth,3)
#best_bandwidth = 0.9

meanshift= MeanShift(bandwidth=best_bandwidth)
cluster_labels_small = meanshift.fit_predict(tmt_scaled_df_small)
print('cluster labels 유형:', np.unique(cluster_labels_small))
```

```python
tmtClusterLblSmall = pd.DataFrame(data = cluster_labels_small, columns = ["cluster"])
#tmtClusterLbl = tmtClusterLbl.reset_index()
#tmtXBig = tmtXDataset.reset_index()
tmtClusterRsltSmall = pd.concat([tmtLiveSmallX.reset_index(drop=True), tmtLiveSmallY.reset_index(drop=True), tmtClusterLblSmall.reset_index(drop=True)],axis = 1 )
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

tmtClusterRsltSmall['cluster']  = cluster_labels_small
centers = meanshift.cluster_centers_
unique_labels = np.unique(cluster_labels_small)
markers=['o', 's', '^', 'x', '*','v','<','>']
plt.figure()
fig, axs = plt.subplots(1,10,figsize=(70,6))
for i, ax in enumerate(axs):
  for label in unique_labels:
      label_cluster = tmtClusterRsltSmall[tmtClusterRsltSmall['cluster']==label]
      center_x_y = centers[label]
      # 군집별로 다른 마커로 산점도 적용
      ax.scatter(x=label_cluster['열매수(개)'], y=label_cluster[label_cluster.columns[i]], edgecolor='k', marker=markers[label] ,alpha=0.5,)

      # 군집별 중심 표현
      ax.scatter(x=center_x_y[0], y=center_x_y[1], s=200, color='gray', alpha=0.9, marker=markers[label])
      ax.scatter(x=center_x_y[0], y=center_x_y[1], s=70, color='k', edgecolor='k', marker='$%d$' % label)
      ax.set_title(label_cluster.columns[i])
  #ax.show()
```
